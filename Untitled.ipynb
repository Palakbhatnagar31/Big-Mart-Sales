```python
pv=data.pivot_table(values='Item_Weight',index='Item_Identifier')
pv.head()
```

```python
def r(x):
    weight=x[0]
    identifier=x[1]
    if pd.isnull(weight):
        return pv['Item_Weight'][pv.index==identifier]
    else:
        return weight
```

```python
data['Item_Weight']=data[['Item_Weight','Item_Identifier']].apply(r,axis=1).astype(float)
```

```python
data[data['Outlet_Size'].isnull()]
```

```python
b=data.pivot_table(values='Outlet_Size',index='Outlet_Type',aggfunc=lambda x:x.mode())
b
```

```python
def g(y):
    size=y[0]
    Type=y[1]
    if pd.isnull(size):
        return b['Outlet_Size'][b.index==Type][0]
    else:
        return size
```

```python
data['Outlet_Size']=data[['Outlet_Size','Outlet_Type']].apply(g,axis=1)
```
